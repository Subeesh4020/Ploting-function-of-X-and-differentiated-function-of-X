# pyplot using f(x) and f'(x)

```python
fx = 0
xaxis = []

for a in range(10):
    fx = ((2 * a**3) + (a**2) - 2) # f(x) = 2x^3 + x^2 - 2
    xaxis.append(fx)
    
fdx = 0
yaxis = []

for a in range(10):
    fdx = ((6 * a**2) + (2 * a))   # f'(x) = 6x^2 + 2x
    yaxis.append(fdx)
```

```python
print(xaxis)
```

```python
print(yaxis)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(xaxis, yaxis)
plt.xlabel("f(x) = 2x^3 + x^2 - 2")
plt.ylabel("f'(x) = 6x^2 + 2x")
plt.title("x - y plot of f(x) and f'(x)")
```
